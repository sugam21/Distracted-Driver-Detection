```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
log = pd.read_csv("log/log.csv")
log.head()
```

```python
plt.plot(log["Epoch"]+1,log["training_accuracy"])
plt.plot(log["Epoch"]+1,log["validation_accuracy"])
plt.legend(["training accu","validation accu"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy Score")
plt.savefig("figures/epoch_vs_accuracy.jpg")
```

```python
plt.plot(log["Epoch"]+1,log["training_loss"])
plt.plot(log["Epoch"]+1,log["validation_loss"])
plt.legend(["training loss","validation loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.savefig("figures/epoch_vs_loss.jpg")
```
